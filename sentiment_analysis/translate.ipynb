```python
import pandas as pd
from translatte import Translator
import re
from tqdm import tqdm
tqdm.pandas()
```

```python
df = pd.read_csv('../data/SUS_Simulador_Aposentadoria_pre_processado.csv')
df.dropna(subset=['Agradeço a sua participação e abro o espaço para que você possa contribuir com alguma crítica, sugestão ou elogio sobre o Simulador de Aposentadoria.'], inplace=True)
df.reset_index(drop=True)
```

```python
def clear_text(text):
    text = re.sub(r'[\n\r\t]', '', text)
    text = text.strip().lower().capitalize()
    return text
```

```python
df["comment_in_english"] = df["Agradeço a sua participação e abro o espaço para que você possa contribuir com alguma crítica, sugestão ou elogio sobre o Simulador de Aposentadoria."].progress_apply(lambda text: Translator.translate(clear_text(text), 'en'))
df
```

```python
df.to_csv('../data/SUS_Simulador_Aposentadoria_comentarios_em_ingles.csv', index=False)
```
