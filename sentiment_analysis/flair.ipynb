```python
import pandas as pd
from flair.models import TextClassifier
from flair.data import Sentence
import numpy as np
import os
from tqdm import tqdm
tqdm.pandas()
```

```python
analyzer = TextClassifier.load('en-sentiment')

def analysis_sentimet(text):
    sentence = Sentence(text)
    analyzer.predict(sentence)
    out = sentence.labels[0]
    if out.value == 'POSITIVE':
        return 1
    else:
        return -1
```

```python
df = pd.read_csv('../data/SUS_Simulador_Aposentadoria_comentarios_em_ingles.csv')
df['comment_in_english'] = df['comment_in_english'].apply(lambda x: 'None' if x is np.nan else x)
df
```

```python
df["flair_result"] = df["comment_in_english"].progress_apply(lambda text: analysis_sentimet(text))
```

```python
df['flair_result'] = pd.to_numeric(df['flair_result'], errors='coerce').astype(int)
```
