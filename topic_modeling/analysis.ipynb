```python
import pandas as pd
```

```python
df_data = pd.read_csv('../data/SUS_Simulador_Aposentadoria_pre_processado.csv')
df_data = df_data.dropna(subset=['clean_text']).reset_index(drop=True)
df_data['Some a pontuação total dos novos valores (X+Y) e multiplique por 2,5.'] = df_data['Some a pontuação total dos novos valores (X+Y) e multiplique por 2,5.'].str.replace(',', '.', regex=False).astype(float)

df_topic_modeling = pd.read_csv('data_topic_modeling/Resumo_Topicos_Dominantes.csv')

df = pd.concat([df_data, df_topic_modeling['dominant_topic']], axis=1)
```

```python
columns = list(df.columns)
columns = columns[3:-6:2] + ['Some a pontuação total dos novos valores (X+Y) e multiplique por 2,5.']
```

```python
grouped_df = df.groupby('dominant_topic')[columns].mean()
    
for topic, group in grouped_df.iterrows():
    print(f"Grupo para dominant_topic = {topic}")
    display(group)
```

```python
grouped_df = df.groupby('dominant_topic')

for topic, group in grouped_df:
    print(f"Grupo para dominant_topic = {topic} ({len(group)} pessoas)")
    group = group[['Agradeço a sua participação e abro o espaço para que você possa contribuir com alguma crítica, sugestão ou elogio sobre o Simulador de Aposentadoria.']]
    display(group)
```
