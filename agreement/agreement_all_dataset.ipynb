```python
from sklearn.metrics import cohen_kappa_score
import statistics
import re
import pandas as pd
import numpy as np
from tqdm import tqdm
import warnings
warnings.filterwarnings("ignore")
```

```python
def contains_number_from_1_to_10(column):
    return bool(re.search(r'\b([1-9]|10)\b', column))

df = pd.read_csv('../data/SUS_Simulador_Aposentadoria_pre_processado.csv')
df = df[[column for column in df.columns[3:-5] if contains_number_from_1_to_10(column)]]
```

```python
kappa = []

for i in tqdm(range(len(df))):    
    answers_i = df.iloc[i].tolist()
    for j in range(i+1, len(df)):
        answers_j = df.iloc[j].tolist()
        if (set(answers_i) == {0}) and (set(answers_j) == {0}):
            k = np.float64(0.0)
        else: 
            k = cohen_kappa_score(answers_i, answers_j)
        kappa.append(k)

mean = statistics.mean(kappa)        
```
